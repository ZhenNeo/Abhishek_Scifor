```python
d = data.groupby(by = 'Continent').sum()

sb.barplot(data = d, x = d.index, y = 'Rank', palette='viridis')
```

```python
column_to_vis = ['2022 Population','2020 Population','2015 Population','2010 Population','2000 Population','1990 Population','1980 Population','1970 Population']

for year in column_to_vis:
    plt.figure(figsize=(10, 6))
    sb.lineplot(data=d, x=d.index, y=year, marker='o', color='skyblue')
    plt.title(f'Total population for {year}')
    plt.xlabel('Country')
    plt.ylabel('Population')
    plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
    plt.grid(True)  # Add grid lines
    plt.tight_layout()  # Adjust layout
    plt.show()
```

```python
total_growth_rate = d['Growth Rate'].sum()

# Calculate percentage of growth rate for each continent
percentage_growth_rate = d['Growth Rate'] / total_growth_rate * 100

# Plot pie chart using Matplotlib
plt.figure(figsize=(8, 8))
plt.pie(percentage_growth_rate, labels=d.index, autopct='%1.1f%%', startangle=140)
plt.title('Total Growth Rate in each continent')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sb.barplot(data=d, x=d.index, y='Area (km²)', palette='viridis')
plt.title('Distribution of Area (km²) Across Continents')
plt.xlabel('Continent')
plt.ylabel('Area in km²')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.grid(True)  # Add grid lines
plt.tight_layout()  # Adjust layout
plt.show()
```

```python
trend = data.iloc[:,5:13].sum().sort_values(ascending=True)

plt.figure(figsize=(10, 6))
sb.lineplot(x=trend.index, y=trend.values, marker='o', color='skyblue')
plt.title('Total Trend Population (1970 to 2020)')
plt.xlabel('Year')
plt.ylabel('Total Population')
plt.grid(True)  # Add grid lines
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()  # Adjust layout
plt.show()
```
